The dataset contains detailed information on **10,000 used cars**, aiming to analyze factors that influence their **resale prices**.

It comprises **10 columns**, capturing a range of vehicle features including make, model, year, engine specifications, usage, and pricing.

This dataset is perfect for building a **used car price prediction model**, performing **feature impact analysis**, and uncovering trends in the second-hand car market.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("car_price_dataset.csv")
df.head()
```

```python
df.info()
```

```python
# Histogram - Price Distribution
plt.figure(figsize=(6, 4))
sns.histplot(df["Price"], bins=30, kde=True, color="blue")
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
# Pie Chart - Fuel Type Distribution
plt.figure(figsize=(6, 4))
fuel_counts = df["Fuel_Type"].value_counts()
plt.pie(fuel_counts, labels=fuel_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Fuel Type Distribution")
plt.show()
```

```python
# 3. Bar Chart - Average Price by Brand (Top 10)
plt.figure(figsize=(8, 5))
brand_prices = df.groupby("Brand")["Price"].mean().sort_values(ascending=False)[:10]
sns.barplot(x=brand_prices.values, y=brand_prices.index, palette="viridis")
plt.title("Average Price by Brand")
plt.xlabel("Average Price")
plt.show()
```

```python
# 4. Stacked Bar Chart - Price Distribution by Fuel Type & Transmission
plt.figure(figsize=(8, 5))
fuel_price = df.groupby(["Fuel_Type", "Transmission"])["Price"].mean().unstack()
fuel_price.plot(kind="bar", stacked=True, colormap="coolwarm")
plt.title("Price by Fuel Type & Transmission")
plt.xlabel("Fuel Type")
plt.ylabel("Average Price")
plt.show()
```

```python

# 5. Box Plot - Price Variation by Transmission Type
plt.figure(figsize=(6, 4))
sns.boxplot(x="Transmission", y="Price", data=df, palette="Set2")
plt.title("Price Variation by Transmission")
plt.show()
```

```python

# 6. Line Plot - Price Trend Over the Years
plt.figure(figsize=(8, 5))
yearly_price = df.groupby("Year")["Price"].mean()
sns.lineplot(x=yearly_price.index, y=yearly_price.values, marker="o", color="red")
plt.title("Price Trend Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

```python

# 7. Scatter Plot - Price vs Mileage
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Mileage", y="Price", data=df, alpha=0.5, color="purple")
plt.title("Price vs Mileage")
plt.xlabel("Mileage (km/l)")
plt.ylabel("Price")
plt.show()
```

```python
# 8. Count Plot - Number of Cars by Transmission Type
plt.figure(figsize=(6, 4))
sns.countplot(x="Transmission", data=df, palette="magma")
plt.title("Count of Cars by Transmission")
plt.xlabel("Transmission")
plt.ylabel("Count")
plt.show()
```

```python

# 9. Pie Chart - Transmission Type Distribution
plt.figure(figsize=(6, 4))
transmission_counts = df["Transmission"].value_counts()
plt.pie(transmission_counts, labels=transmission_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Transmission Type Distribution")
plt.show()
```

```python

# 10. Box Plot - Price by Owner Count
plt.figure(figsize=(6, 4))
sns.boxplot(x="Owner_Count", y="Price", data=df, palette="coolwarm")
plt.title("Price Variation by Owner Count")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import pandas as pd



data_encoded = pd.get_dummies(df, drop_first=True)


X = data_encoded.drop('Price', axis=1, errors='ignore') 
y = data_encoded['Price']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Model evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")

plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Car Prices")
plt.show()
```
